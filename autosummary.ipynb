```python
from typing_extensions import Text
from bs4 import BeautifulSoup
import re
html_content = html

soup = BeautifulSoup(html_content, 'html.parser')

output_text = ""
# Haal alle divjes met .pl-example class eruit, (blokjes met gele achtergrond)
for elem in soup.select(".pl-example"):
    elem.extract()

# Dit is vooral om de verschllende HTML elementen om te zetten naar Markdown achtige output.
for element in soup.find_all(['p', 'ul', 'li', 'h2', 'h4', 'b']):
    if element.name == 'p':
        output_text += element.get_text() + "``"
    elif element.name == 'ul':
        continue
    elif element.name == 'li':
        output_text += "- " + element.get_text() + ""
    elif element.name == 'b':
        output_text += "*" + element.get_text() + "* "
    elif element.name == 'h2':
        output_text += "## " + element.get_text() + "\n\n"
    elif element.name == 'h4':
        output_text += "#### " + element.get_text() + "\n\n"

text =  re.sub(r'\s+', ' ', output_text).strip()
text = text.replace('``', '\n\n')

text
```

```python
# Splitting de tekst on '##'
sections = text.split('##')

# En plakken de tekst achter ## er weer aan vast- en plaatsen hem terug.
sections = ['##' + section if index != 0 else section for index, section in enumerate(sections) if section.strip() != '']
sections
```
